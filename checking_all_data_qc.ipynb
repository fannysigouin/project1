# Checking occurrance numbers

Becuase occurrance ids might have several rows of associated data, we are checking how often this happens.

```python
# Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Read in datasets and merge them into a single df
# Create empty df to store data
all_data = pd.read_csv('resources/mci_all.csv')
all_data.head()
```

```python
no_duplicates_df = all_data.drop_duplicates(['EVENT_UNIQUE_ID', 'OFFENCE'])

# confirm that there are no duplicates of the same event and offence
no_duplicates_df[['EVENT_UNIQUE_ID', 'OFFENCE']].value_counts(ascending=False)
```

```python
no_duplicates_df.to_csv(Path('resources', 'mci_no_duplicates_all.csv'), index = False)
```


---

## Check occurrance year counts

```python
# Check counts of occurance year

df_by_year = (no_duplicates_df
    .groupby('OCC_YEAR')
    .size()
)
df_by_year
```

```python
# Keep only 2014 and over
over_2014_df = (no_duplicates_df
    .loc[no_duplicates_df['OCC_YEAR'] >= 2014]
)
over_2014_df.groupby('OCC_YEAR').size()
```

```python
# save data from 2014 and over
over_2014_df.to_csv(Path('resources', 'mci_no_duplicates_2014_and_over.csv'), index=False)
```
