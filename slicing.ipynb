```python
# dependencies
import pandas as pd
```

```python
# read in data
all_data = pd.read_csv('resources/mci_all.csv')
all_data.head()
```

```python
# get data from 2018, 2020 and 2022 and output to separate files
years = [2018, 2020, 2022]

for year in years:
    this_year = all_data.loc[all_data['OCC_YEAR']==year]
    this_year.to_csv('resources/mci_' + str(year) + '.csv', header=True, index=False)
```
